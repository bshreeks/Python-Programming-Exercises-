```python
# diff_chaos.py
# program to illustrate chaos problem with algebraically equivalent equations

def main():
    print("This program illustrates the chaotic behavior of algebraically equivalent equations")
    
    x = eval(input("Enter a number between 0 and 1: "))
    
    for i in range(100):
        x = 3.9 * x * (1-x)
        print(x)
    print("\n")
    
    for i in range(100):
        x = 3.9 * (x - x * x)
        print(x)
    print("\n")
        
    for i in range(100):
        x = 3.9 * x - 3.9 * x * x
        print(x)
main()
```
