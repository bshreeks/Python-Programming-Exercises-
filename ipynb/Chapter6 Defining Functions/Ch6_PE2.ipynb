```python
#ants_go_march.py
#Program to print the lyrics of the song 'the ants go marching' for ten verses

#define the first verse
def frt_verse(num):
    print("The ants go marching " + num + " by " + num +"," + " hurrah! hurrah!") 

#define the second verse
def sec_verse(num, act):
    print("The ants go marching " + num + " by " + num)
    print("The little one stops to " + act)
    print("And They all go marching down...")
    print("In the ground...")
    print("To get out....")
    print("Of the rain.")
    print("Boom! Boom! Boom!")          
    #lyrics = frt_verse() + num + " by " + num +"," + " hurrah! hurrah!\n" + frt_verse() + num + " by " + num +"," + " hurrah! hurrah!\n" + frt_verse()+ num + " by " + num +",\n" + "The little one stops to " + act +",\n" + "And They all go marching down...\n" + "In the ground...\n" + "To get out...\n" + "Of the rain.\n" + "Boom! Boom! Boom!"

def main():
    print("This program prints the lyrics of the song 'The ants go marching' for ten verses\n")
    num = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten"]
    act = ["suck the thumb", " tie his shoe", "climb a tree", "shut the door", "take a dive", "pick up sticks", "pray to heaven", "check the gate", "check the time", "say The End!"]
    for n, ac in zip(num, act):
        frt_verse(n)
        frt_verse(n)
        sec_verse(n, ac)
        
main()
```
